```python
import heapq

class TSP_AStar:
    def __init__(self, matrix):
        self.matrix = matrix
        self.n = len(matrix)

    def heuristic(self, current_node, unvisited):
        """
        Hàm h(n): Ước lượng chi phí từ các nút chưa thăm để hoàn thành chu trình.
        Ở đây dùng phương pháp đơn giản: (số nút chưa thăm + 1) * cạnh nhỏ nhất của đồ thị.
        """
        if not unvisited:
            return self.matrix[current_node][0]


        min_edge = min([min([row for row in line if row > 0]) for line in self.matrix])
        return (len(unvisited) + 1) * min_edge

    def solve(self):

        start_node = 0
        unvisited_start = tuple(range(1, self.n))


        open_list = []
        initial_h = self.heuristic(start_node, unvisited_start)
        heapq.heappush(open_list, (initial_h, start_node, unvisited_start, 0, [start_node]))

        while open_list:
            f, current, unvisited, g, path = heapq.heappop(open_list)


            if not unvisited:
                total_cost = g + self.matrix[current][0]
                final_path = path + [0]
                return final_path, total_cost


            for i, next_node in enumerate(unvisited):
                new_g = g + self.matrix[current][next_node]
                new_unvisited = unvisited[:i] + unvisited[i+1:]
                new_f = new_g + self.heuristic(next_node, new_unvisited)

                heapq.heappush(open_list, (new_f, next_node, new_unvisited, new_g, path + [next_node]))

        return None, float('inf')


matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp_solver = TSP_AStar(matrix)
path, cost = tsp_solver.solve()

print(f"Đường đi tối ưu: {path}")
print(f"Tổng chi phí: {cost}")
```
