```python
#cau13
import re
def merge_and_sum(a, b):
    res = []
    min_len = min(len(a), len(b))

    # Sum corresponding elements
    for i in range(min_len):
        res.append(a[i] + b[i])

    # Append remaining elements if one array is longer
    if len(a) > len(b):
        res.extend(a[min_len:])
    else:
        res.extend(b[min_len:])

    return res

# Example based on your request
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
result = merge_and_sum(a, b)

print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Mảng kết quả: {result}")
```

```python
import random

def create_matrix(m, n):
    # a. Tạo mảng a chứa các số nguyên ngẫu nhiên
    return [[random.randint(0, 50) for _ in range(n)] for _ in range(m)]

def print_row_k(a, k):
    # b. Xuất các phần tử thuộc dòng k
    if 0 <= k < len(a):
        print(f"Dòng {k}: {a[k]}")

def print_col_k(a, k):
    # c. Xuất các phần tử thuộc cột k
    if len(a) > 0 and 0 <= k < len(a[0]):
        col = [row[k] for row in a]
        print(f"Cột {k}: {col}")

def find_row_max_sum(a):
    # d. Tìm dòng có tổng lớn nhất
    max_sum = -float('inf')
    row_idx = -1
    for i, row in enumerate(a):
        s = sum(row)
        if s > max_sum:
            max_sum = s
            row_idx = i
    print(f"Dòng có tổng lớn nhất là dòng {row_idx} với tổng {max_sum}")

def find_col_min_product(a):
    # e. Tìm cột có tích nhỏ nhất
    if not a or not a[0]: return
    min_prod = float('inf')
    col_idx = -1
    for j in range(len(a[0])):
        prod = 1
        for i in range(len(a)):
            prod *= a[i][j]
        if prod < min_prod:
            min_prod = prod
            col_idx = j
    print(f"Cột có tích nhỏ nhất là cột {col_idx} với tích {min_prod}")

def print_even_row_odd_col(a):
    # f. Xuất ra các phần tử thuộc dòng chẵn và cột lẻ
    res = []
    for i in range(0, len(a), 2):
        for j in range(1, len(a[0]), 2):
            res.append(a[i][j])
    print(f"Phần tử dòng chẵn, cột lẻ: {res}")

def avg_even_in_odd_rows(a):
    # g. Trung bình cộng các phần tử chẵn thuộc dòng lẻ
    evens = []
    for i in range(1, len(a), 2):
        for val in a[i]:
            if val % 2 == 0:
                evens.append(val)
    avg = sum(evens) / len(evens) if evens else 0
    print(f"TBC phần tử chẵn thuộc dòng lẻ: {avg}")

def avg_boundary(a):
    # h. Trung bình cộng các phần tử thuộc biên
    m, n = len(a), len(a[0])
    boundary = []
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                boundary.append(a[i][j])
    avg = sum(boundary) / len(boundary) if boundary else 0
    print(f"TBC phần tử biên: {avg}")

def avg_product_inside(a):
    # i. Trung bình tích các phần tử không thuộc biên (Trung bình nhân)
    import math
    m, n = len(a), len(a[0])
    inside = []
    for i in range(1, m-1):
        for j in range(1, n-1):
            inside.append(a[i][j])
    if not inside:
        print("Không có phần tử bên trong.")
        return
    product = math.prod(inside)
    avg_prod = product ** (1/len(inside))
    print(f"Trung bình tích (Trung bình nhân) phần tử bên trong: {avg_prod}")

# Chạy thử
m, n = 4, 5
matrix = create_matrix(m, n)
for row in matrix: print(row)
print_row_k(matrix, 1)
print_col_k(matrix, 2)
find_row_max_sum(matrix)
find_col_min_product(matrix)
print_even_row_odd_col(matrix)
avg_even_in_odd_rows(matrix)
avg_boundary(matrix)
avg_product_inside(matrix)
```

```python
from datetime import datetime

# a. Xây dựng cấu trúc SINHVIEN mô tả một sinh viên
class SINHVIEN:
    def __init__(self, ma_sv, ten, nam_sinh, diem_tb):
        self.ma_sv = str(ma_sv)[:10]  # Tối đa 10 ký tự
        self.ten = str(ten)[:20]      # Tối đa 20 ký tự
        self.nam_sinh = int(nam_sinh)
        self.diem_tb = float(diem_tb)

    def __repr__(self):
        return f"[Mã: {self.ma_sv}, Tên: {self.ten}, Năm sinh: {self.nam_sinh}, ĐTB: {self.diem_tb}]"

# c. Hàm cho biết có bao sinh viên đủ điều kiện lên lớp (ĐTB >= 5)
def dem_sv_len_lop(danh_sach):
    count = sum(1 for sv in danh_sach if sv.diem_tb >= 5)
    return count

# d. Xuất các sinh viên đủ 20 tuổi
def xuat_sv_20_tuoi(danh_sach):
    current_year = datetime.now().year
    print("Danh sách sinh viên đủ 20 tuổi:")
    for sv in danh_sach:
        if current_year - sv.nam_sinh == 20:
            print(sv)

# e. Đếm số sinh viên học hệ đại học (Mã có 'DH' ở vị trí 2, 3 - index 2:4)
def dem_sv_dai_hoc(danh_sach):
    count = sum(1 for sv in danh_sach if len(sv.ma_sv) >= 4 and sv.ma_sv[2:4] == 'DH')
    return count

# f. Đếm sinh viên có tên «Lan»
def dem_ten_lan(danh_sach):
    count = 0
    for sv in danh_sach:
        ten_tach = sv.ten.strip().split()
        if ten_tach and ten_tach[-1].lower() == "lan":
            count += 1
    return count

# g. Đếm sinh viên có họ «Phan»
def dem_ho_phan(danh_sach):
    count = 0
    for sv in danh_sach:
        ten_tach = sv.ten.strip().split()
        if ten_tach and ten_tach[0].lower() == "phan":
            count += 1
    return count

# b. Khởi tạo mảng n sinh viên (Dữ liệu mẫu)
danh_sach_sv = [
    SINHVIEN("01DH0001", "Phan Thanh Lan", 2004, 7.5),
    SINHVIEN("02DH0002", "Nguyen Van A", 2003, 4.2),
    SINHVIEN("03CD1234", "Phan Van B", 2004, 6.0),
    SINHVIEN("04DH5678", "Tran Thi Lan", 2005, 8.5),
    SINHVIEN("05DH1111", "Le Van Phan", 2004, 3.0)
]

# Chạy thử các hàm
print(f"Số SV đủ điều kiện lên lớp: {dem_sv_len_lop(danh_sach_sv)}")
xuat_sv_20_tuoi(danh_sach_sv)
print(f"Số SV hệ Đại học: {dem_sv_dai_hoc(danh_sach_sv)}")
print(f"Số SV tên Lan: {dem_ten_lan(danh_sach_sv)}")
print(f"Số SV họ Phan: {dem_ho_phan(danh_sach_sv)}")
```
