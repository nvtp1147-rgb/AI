```python
#1.
import re
def my_function(arr1, arr2):
    result = []
    len1 = len(arr1)
    len2 = len(arr2)

    min_len = min(len1, len2)
    for i in range(min_len):
        result.append(min(arr1[i], arr2[i]))


    if len1 > min_len:
        result.extend(arr1[min_len:])


    if len2 > min_len:
        result.extend(arr2[min_len:])

    return result


a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

result_array = my_function(a, b)
print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Mảng kết quả: {result_array}")
```

```python

#a
def tinh_tong_tam_giac_tren_cheo_phu(a, n):
    """a. Tính tổng các phần tử thuộc tam giác trên đường chéo phụ (kể cả đường chéo phụ)"""
    tong = 0
    for i in range(n):
        for j in range(n - i):
            tong += a[i][j]
    return tong
#b
def chuyen_am_thanh_tri_tuyet_doi(a, m, n):
    """b. Chuyển các phần tử âm thành trị tuyệt đối"""
    for i in range(m):
        for j in range(n):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
    return a

#c
def thay_chan_bang_x(a, m, n, x):
    """c. Thay các phần tử chẵn bằng số nguyên x cho trước"""
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    return a

#d
def kiem_tra_toan_chan(a, m, n):
    """d. Kiểm tra a có toàn chẵn không?"""
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 != 0:
                return False
    return True

#e
def kiem_tra_doi_xung(a, n):
    """e. Kiểm tra ma trận vuông a có đối xứng không (a[i][j] == a[j][i])"""
    for i in range(n):
        for j in range(i + 1, n):
            if a[i][j] != a[j][i]:
                return False
    return True

#f
def kiem_tra_cheo_chinh_tang_dan(a, n):
    """f. Kiểm tra đường chéo chính có tăng dần không?"""
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

#g
def xuat_tam_giac_duoi_cheo_phu(a, n):
    """g. Xuất các phần tử thuộc tam giác dưới đường chéo phụ (kể cả đường chéo phụ)"""
    ket_qua = []
    for i in range(n):
        for j in range(n - 1 - i, n):
            ket_qua.append(a[i][j])
    return ket_qua

#h
def kiem_tra_cheo_phu_giam_dan(a, n):
    """h. Kiểm tra đường chéo phụ có giảm dần không?"""
    # Chỉ số đường chéo phụ là a[i][n - 1 - i]
    for i in range(n - 1):
        if a[i][n - 1 - i] <= a[i+1][n - 2 - i]:
            return False
    return True

# --- Ví dụ chạy thử ---
matrix = [
    [1, 2, 3],
    [2, 5, 6],
    [3, 6, 9]
]
n = 3

print("Tổng tam giác trên chéo phụ:", tinh_tong_tam_giac_tren_cheo_phu(matrix, n))
print("Ma trận có đối xứng không?", kiem_tra_doi_xung(matrix, n))
print("Đường chéo chính có tăng dần?", kiem_tra_cheo_chinh_tang_dan(matrix, n))
```
